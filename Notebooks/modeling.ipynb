```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = "../data/raw/customer_data.csv"
df = pd.read_csv(filepath)
df
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
train_vl_df,test_df = train_test_split(df,test_size= 0.2,random_state= 42)
train_df,val_df = train_test_split(train_vl_df,test_size = 0.2,random_state = 45)
```

```python
train_df
```

```python
test_df
```

```python
val_df
```

```python
input_cols = list(train_df.columns)[1:-1]
traget_col = 'Churn'
```

```python
input_cols
```

```python
traget_col
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[traget_col].copy()
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[traget_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_targets = test_df[traget_col].copy()
```

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes(include=np.object_).columns.tolist()
```

```python
train_inputs[numeric_cols].describe()
```

```python
train_inputs[categorical_cols].nunique()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
```

```python
imputer.fit(df[numeric_cols])
```

```python
list(imputer.statistics_)
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
train_inputs.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[numeric_cols])
```

```python
print('min')
list(scaler.data_min_)
```

```python
print('max')
list(scaler.data_max_)
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].describe()
```

```python
df[categorical_cols].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
encode.fit(df[categorical_cols])
```

```python
encode.categories_
```

```python
encode_col = list(encode.get_feature_names_out(categorical_cols))
print(encode_col)
```

```python
train_inputs[encode_col] = encode.transform(train_inputs[categorical_cols])
val_inputs[encode_col] = encode.transform(val_inputs[categorical_cols])
test_inputs[encode_col] = encode.transform(test_inputs[categorical_cols])
```
